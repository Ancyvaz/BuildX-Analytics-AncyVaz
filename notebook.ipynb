```python
import pandas as pd
df = pd.read_csv("Placement_Data_Full_Class.csv")
```

```python
# Print structural metrics to verify successful intake
print(f"Dataset Loaded Successfully! Rows: {df.shape[0]}, Columns: {df.shape[1]}")
print("\n--- FIRST 5 RECORDS ---")
print(df.head())
```

```python

# STEP 2: MISSING VALUE HANDLING & DUPLICATES
# Print structural evaluation of null spaces
print("\n--- NULL COUNT BEFORE TREATMENT ---")
print(df.isnull().sum())

#  In this specific dataset, missing values in the 'salary' column
# correspond explicitly to students whose 'placement_status' is 'Not Placed'.
# Therefore, we fill these null fields with a logical numeric 0 value.
df['salary'] = df['salary'].fillna(0)

# Drop accidental row duplicates to guarantee operational data integrity
initial_rows = df.shape[0]
df = df.drop_duplicates()
print(f"\nRemoved {initial_rows - df.shape[0]} duplicate rows.")
```

```python
# Clean text structural anomalies and align casing standardizations
for col in ['degree_t', 'workex', 'status', 'specialisation']:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.title()

# Fix incorrect decimal floating artifacts by converting financial column to integer values
df['salary'] = df['salary'].astype(int)

# FEATURE ENGINEERING: Create an essential derived analytical metric
# 'academic_gap' maps the difference between High School performance and Undergrad performance
df['academic_gap'] = df['hsc_p'] - df['degree_p']

# Rename columns cleanly for professional database compatibility
df.rename(columns={
    'degree_p': 'ug_percentage',
    'degree_t': 'ug_stream',
    'status': 'placement_status'
}, inplace=True)

# Export the cleaned artifact for SQL/Power BI stages
df.to_csv("cleaned_dataset.csv", index=False)
print("\nExport Complete! File saved cleanly as 'cleaned_dataset.csv'")
```
